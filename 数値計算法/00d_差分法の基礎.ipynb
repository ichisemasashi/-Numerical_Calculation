## 偏導関数の差分近似

2つの独立変数$x$と$y$による関数を考える。

変数$(x, y)$を$x$-$y$平面にとって、この$x$、$y$方向に均一幅$h$、$k$の格子座標$(i, j)$を考える。

この$i$-$j$座標に垂直な軸$u$を取る。

$x$と$y$を変数としてもつ関数$u(x, y)$の値は$u$軸の値を取り、$u=u(x,y)$は局面で表される。

格子座標$i$-$j$上の関数値$u(x,y)$を$u_{i,j}$と表記する。

ここで、変数$y$が$y = j$である場合を考えると、関数$u(x,y)$は変数$x$のみの関数になり、その図形は$x$-$u$座標上の曲線になる。

よって、$x=i$における$x$軸方向の勾配、すなわち導関数は、その$x$方向の近傍の関数値で近似できる。

ただし、関数$u(x,y)$は$x$と$y$という2つの独立変数を持つので、導関数は偏導関数となる。

この偏導関数は、$h$が微小であれば、$u_{i,j}$とその前後の格子点上の関数値$u_{i-1,j}$、$u_{i+1,j}$を用いて次のように書くことができる。

$$
\begin{array}{}
\left( \cfrac{\partial u}{\partial x} \right)_{i,j} & = & \cfrac{u_{i,j} - u_{i-1,j}}{h} & \cdots 後退差分近似\\
\left( \cfrac{\partial u}{\partial x} \right)_{i,j} & = & \cfrac{u_{i+1,j} - u_{i,j}}{h} & \cdots 前進差分近似\\
\left( \cfrac{\partial u}{\partial x} \right)_{i,j} & = & \cfrac{u_{i+1,j} - u_{i-1,j}}{2h} & \cdots 中心差分近似
\end{array}
$$

偏導関数$(\partial u/\partial x)_{i,j}$とは、格子座標$(i,j)$における関数$u$の$x$方向への接線の勾配である。

2次の偏導関数$(\partial ^2 u/\partial x^2)_{i,j}$は、1次偏導関数の増分を$h$で割ったものだから、$(i,j)$と$(i-1,j)$の前進差分を用いて

$$
\begin{array}{}
\left( \cfrac{\partial ^2 u}{\partial x^2} \right)_{i,j} & = & \cfrac{(\partial u/\partial x)_{i,j} - (\partial u/\partial x)_{i-1,j}}{h}\\
  & = & \cfrac{(u_{i+1,j} - u_{i,j}) - (u_{i,j} - u_{i-1,j})}{h^2}\\
  & = & \cfrac{u_{i+1,j} - 2u_{i,j} + u_{i-1,j}}{h^2}\\
\end{array}
$$

※ ↑の式では、後退差分と前進差分をミックスしている。前進差分を用いる場合、$(i+1,j)$と$(i,j)$の差分になるのでは?
$$
\begin{array}{}
\left( \cfrac{\partial ^2 u}{\partial x^2} \right)_{i,j} & = & \cfrac{(\partial u/\partial x)_{i+1,j} - (\partial u/\partial x)_{i,j}}{h}\\
  & = & \cfrac{(u_{i+2,j} - u_{i+1,j}) - (u_{i+1,j} - u_{i,j})}{h^2}\\
  & = & \cfrac{u_{i+2,j} - 2u_{i+1,j} + u_{i,j}}{h^2}\\
\end{array}
$$

関数$u$は$x$と$y$の関数であるので、変数$y$についても同様に偏導関数は下のようになる。

$$\left( \cfrac{\partial u}{\partial x} \right)_{i,j} = \cfrac{u_{i,j+1} - u_{i,j-1}}{2k}$$

$$\left( \cfrac{\partial ^2 u}{\partial x^2} \right)_{i,j} = \cfrac{u_{i,j+1} - u_{i,j-1}}{2k}$$

このような変換法により、偏微分方程式は差分方程式に変換できる。

偏微分方程式は、放物型、双曲型、楕円型に分類される。
