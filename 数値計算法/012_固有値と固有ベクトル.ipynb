# 固有値と固有ベクトル

多くの振動体が結合しているとき、その系にはいくつかの共振周波数が生まれる。その値は固有値とよばれ、個々に立てた方程式群の係数行列がもつ特有の値になる。

## 固有値について

次の同次方程式を考える。

$$
\left.
\begin{array}{}
   3x_1 & - x_2 & = \lambda x_1\\
   -x_1 & + x_2 & = 2\lambda x_2
\end{array}
\right\}
\tag{1}$$

この式を行列式で表現すると次のようになり、行列$\mathbb{B}$、$\mathbb{C}$とも対角線に対して対称な行列となり、これを対称行列とよぶ。

$\mathbb{Bx} = \lambda \mathbb{Cx}\tag{2}$

$\mathbb{B} =
\left[
\begin{array}{cc}
 3 & -1\\
-1 &  1
\end{array}
\right]
$、
$\mathbb{C} =
\left[
\begin{array}{cc}
 1 & 0\\
 0 & 2
\end{array}
\right]
$、
$\mathbb{x} =
\left[
\begin{array}{c}
 x_1 \\
 x_2 
\end{array}
\right]
$


さて、$(2)$の両辺に左から$\mathbb{C}$の逆行列$\mathbb{C}^{-1}$を乗算すると

$$\mathbb{C}^{-1}\mathbb{Bx} = \lambda \mathbb{x}$$

となる。ここで、$\mathbb{C}^{-1}\mathbb{B}$を$\mathbb{A}$と置くと

$\mathbb{A} = \mathbb{C}^{-1}\mathbb{B} =
\left[\begin{array}{cc}
 1 & 0\\
 0 & 1/2
\end{array}\right]
\left[\begin{array}{cc}
  3 & -1\\
 -1 &  1
\end{array}\right]
=
\left[\begin{array}{cc}
 3    & -1\\
 -1/2 & 1/2
\end{array}\right]
$

となり、

$$\mathbb{Ax} = \lambda \mathbb{x}$$

である。この同次方程式を満足する$\lambda$を、行列$\mathbb{A}$の固有値とよぶ。
それを求めるために、単位行列$\mathbb{I}$を用いて

$$(\mathbb{A} - \lambda \mathbb{I}) \mathbb{x} = 0 \tag{3}$$

$\mathbb{x}$は式$(1)$の解であり、値を持つはずだから、各要素は$0$ではない。よって、上式の係数行列の行列式は$0$でなければならない。

$$|\mathbb{A} - \lambda \mathbb{I}| = 0$$

$$
\begin{array}{ccc}
|\mathbb{A} - \lambda \mathbb{I}| &=&
\left|
\left[\begin{array}{cc}
   3  & -1\\
 -1/2 & 1/2
\end{array}\right]
- \lambda
\left[\begin{array}{cc}
  1 & 0\\
  0 & 1
\end{array}\right]
\right|\\
  &=&
\left| \begin{array}{cc}
  3 - \lambda & -1\\
  -1/2        & 1/2 - \lambda
\end{array} \right|\\
  &=& (3 - \lambda)(1/2 - \lambda) - (-1)(-1/2)\\
  &=& 3/2 - (3 + 1/2)\lambda + \lambda ^2 - 1/2\\
  &=& \lambda ^2 - (7/2) \lambda + 1\\
  &=& 0
\end{array}
$$

よって、2根$\lambda _1$、$\lambda _2$が求められ、それぞれに対して$x_1$、$x_2$の値の比が求められる。

このように、行列$\mathbb{A}$に関する方程式$(3)$が与えられたとき、その方程式を固有方程式とよび、その根$\lambda _1$、$\lambda _2$を行列$\mathbb{A}$の固有値、その固有値に対応する解$\mathbb{x}$を行列$\mathbb{A}$の固有ベクトルとよぶ。



## 固有方程式

行列$\mathbb{B}$、$\mathbb{C}$は対称行列であっても、行列$\mathbb{A}$は一般的には対称行列にならない。

しかし、数学的扱いは$\mathbb{A}$が対称行列であることが望ましいので、式$(3)$の行列$\mathbb{A}$が対称行列になるような変換をしよう。

$\mathbb{C} = \mathbb{C}^{1/2} \mathbb{C}^{1/2}$ 、また、$\mathbb{C}^{1/2}\mathbb{C}^{-1/2} = \mathbb{I}$なので、式$(2)$は次のように変形できる。

$$\mathbb{B}(\mathbb{C}^{-1/2}\mathbb{C}^{1/2})\mathbb{x} = \lambda (\mathbb{C}^{1/2}\mathbb{C}^{1/2})\mathbb{x}$$

この両辺に左から逆行列$\mathbb{C}^{-1/2}$をかけると

$$\mathbb{C}^{-1/2}\mathbb{BC}^{-1/2}\mathbb{C}^{1/2}\mathbb{x} = \lambda \mathbb{C}^{1/2}\mathbb{x}\tag{4}$$

ここで、

$\mathbb{C}^{-1/2}\mathbb{BC}^{-1/2} = \mathbb{A}$、$\mathbb{C}^{1/2}\mathbb{x} = \mathbb{y}$

と置くと、

$$\mathbb{Ay} = \lambda \mathbb{y}$$

となる。ただし、$\mathbb{C}$が対称行列の場合に限られる。
