# 固有値と固有ベクトル

多くの振動体が結合しているとき、その系にはいくつかの共振周波数が生まれる。その値は固有値とよばれ、個々に立てた方程式群の係数行列がもつ特有の値になる。

## 固有値について

次の同次方程式を考える。

$$
\left.
\begin{array}{}
   3x_1 & - x_2 & = \lambda x_1\\
   -x_1 & + x_2 & = 2\lambda x_2
\end{array}
\right\}
\tag{1}$$

この式を行列式で表現すると次のようになり、行列$\mathbb{B}$、$\mathbb{C}$とも対角線に対して対称な行列となり、これを**対称行列**とよぶ。

$\mathbb{Bx} = \lambda \mathbb{Cx}\tag{2}$

$\mathbb{B} =
\left[
\begin{array}{cc}
 3 & -1\\
-1 &  1
\end{array}
\right]
$、
$\mathbb{C} =
\left[
\begin{array}{cc}
 1 & 0\\
 0 & 2
\end{array}
\right]
$、
$\mathbb{x} =
\left[
\begin{array}{c}
 x_1 \\
 x_2 
\end{array}
\right]
$


さて、$(2)$の両辺に左から$\mathbb{C}$の逆行列$\mathbb{C}^{-1}$を乗算すると

$$\mathbb{C}^{-1}\mathbb{Bx} = \lambda \mathbb{x}$$

となる。ここで、$\mathbb{C}^{-1}\mathbb{B}$を$\mathbb{A}$と置くと

$\mathbb{A} = \mathbb{C}^{-1}\mathbb{B} =
\left[\begin{array}{cc}
 1 & 0\\
 0 & 1/2
\end{array}\right]
\left[\begin{array}{cc}
  3 & -1\\
 -1 &  1
\end{array}\right]
=
\left[\begin{array}{cc}
 3    & -1\\
 -1/2 & 1/2
\end{array}\right]
$

となり、

$$\mathbb{Ax} = \lambda \mathbb{x}$$

である。この同次方程式を満足する$\lambda$を、行列$\mathbb{A}$の**固有値**とよぶ。
それを求めるために、単位行列$\mathbb{I}$を用いて

$$(\mathbb{A} - \lambda \mathbb{I}) \mathbb{x} = 0 \tag{3}$$

$\mathbb{x}$は式$(1)$の解であり、値を持つはずだから、各要素は$0$ではない。よって、上式の係数行列の行列式は$0$でなければならない。

$$|\mathbb{A} - \lambda \mathbb{I}| = 0$$

$$
\begin{array}{ccl}
|\mathbb{A} - \lambda \mathbb{I}| &=&
\left|
\left[\begin{array}{cc}
   3  & -1\\
 -1/2 & 1/2
\end{array}\right]
- \lambda
\left[\begin{array}{cc}
  1 & 0\\
  0 & 1
\end{array}\right]
\right|\\
  &=&
\left| \begin{array}{cc}
  3 - \lambda & -1\\
  -1/2        & 1/2 - \lambda
\end{array} \right|\\
  &=& (3 - \lambda)(1/2 - \lambda) - (-1)(-1/2)\\
  &=& 3/2 - (3 + 1/2)\lambda + \lambda ^2 - 1/2\\
  &=& \lambda ^2 - (7/2) \lambda + 1\\
  &=& 0
\end{array}
$$

よって、2根$\lambda _1$、$\lambda _2$が求められ、それぞれに対して$x_1$、$x_2$の値の比が求められる。

このように、行列$\mathbb{A}$に関する方程式$(3)$が与えられたとき、その方程式を**固有方程式**とよび、その根$\lambda _1$、$\lambda _2$を**行列$\mathbb{A}$の固有値**、その固有値に対応する解$\mathbb{x}$を行列$\mathbb{A}$の**固有ベクトル**とよぶ。



>図のように、質量が$m$と$2m$の2質点が、ばね定数が$2k$と$k$のばねで直列に支えられた自由振動系において、固有振動数と各質点の最大振幅の比を求める。

![07_02.jpg](07_02.jpg)

各質点の変位を$u_1$、$u_2$とする。各質点の力の平衡方程式は(外力=質量$\times$加速度)で表されるから、

$$
\left.
\begin{array}{rcl}
-2ku_1 + k(u_2 - u_1) & = & m \dfrac{d^2u_1}{dt^2}\\
-k(u_2 - u_1) & = & 2m \dfrac{d^2u_2}{dt^2}
\end{array}
\right\}
$$

各質点の最大振幅が$x_1$、$x_2$で、角周波数が$\omega$の振動をするものと仮定して、

$$ u_1 = x_1 \sin (\omega t)、 u_2 = x_2 \sin (\omega t)$$

を平衡方程式に代入すると

$$
\left.
\begin{array}{rcl}
-2x_1 + (x_2 - x_1) & = & - \dfrac{m \omega ^2}{k}x_1\\
-(x_2 - x_1) & = & -2 \dfrac{m \omega ^2}{k}x_2
\end{array}
\right\}
$$

となる。よって、無次元化された振動数として、$\lambda = m \omega^2 / k$とおくと、$(1)$と同じ2元連立1次方程式になる。よって、この場合は、**固有値の平方根は系の無次元化共振周波数**を表わし、質点の**最大振幅の比が固有ベクトル**に現れてくる。


> $m$、$2m$、$3m$という3つの質点が、ばね定数$3k$、$2k$、$k$というばねで天井から順次直列につり下げられている振動系において、共振周波数とそのときの各質点の最大振幅比を求めよ。



各質点での力の平衡方程式を立てると、

$$
\left.
\begin{array}{rcl}
-3ku_1 + 2k(u_2 - u_1) & = & m \dfrac{d^2u_1}{dt^2}\\
-2k(u_2 - u_1) + k(u_3 - u_2) & = & 2m \dfrac{d^2u_2}{dt^2}\\
-k(u_3 - u_2) & = & 3m \dfrac{d^2u_3}{dt^2}
\end{array}
\right\}
$$

ここで、
$$ u_1 = x_1 \sin (\omega t) 、 u_2 = x_2 \sin (\omega t) 、 u_3 = x_3 \sin (\omega t) 、$$
を代入し、$\lambda = m \omega ^2 / k$とおくと、次のようになる。

$$\mathbb{Bx} = \lambda \mathbb{Cx}$$
$$\mathbb{B} =
\left(
\begin{array}{rrr}
5 & -1 & 0\\
-2 & 3 & -1\\
0 & -1 & 1
\end{array}
\right)
、\mathbb{C} = 
\left(
\begin{array}{rrr}
1 & 0 & 0\\
0 & 2 & 0\\
0 & 0 & 3
\end{array}
\right)
、\mathbb{x} = 
\left(
\begin{array}{r}
x_1\\
x_2\\
x_3
\end{array}
\right)
$$

ところで、$\mathbb{C}$の逆行列$\mathbb{C}^{-1}$を求めると次のようになる。

$$
\mathbb{C}^{-1} =
\left(
\begin{array}{rrr}
1 & 0 & 0\\
0 & 2 & 0\\
0 & 0 & 3
\end{array}
\right)^{-1} = 
\left(
\begin{array}{rrr}
1 & 0 & 0\\
0 & 1/2 & 0\\
0 & 0 & 1/3
\end{array}
\right)
$$

よって、$|\mathbb{C}^{-1}\mathbb{B} - \lambda \mathbb{I}| = 0$を作ると、

$$
\left|
\left(
\begin{array}{rrr}
1 & 0 & 0\\
0 & 1/2 & 0\\
0 & 0 & 1/3
\end{array}
\right)
\left(
\begin{array}{rrr}
5 & -2 & 0\\
-2 & 3 & -1\\
0 & -1 & 1
\end{array}
\right)
- \lambda
\left(
\begin{array}{rrr}
1 & 0 & 0\\
0 & 1 & 0\\
0 & 0 & 1
\end{array}
\right)
\right| = 
\left|
\begin{array}{rrr}
\lambda - 5 & 2 & 0\\
1 & \lambda - 3/2 & 1/2\\
0 & 1/3 & \lambda - 1/3
\end{array}
\right| = \lambda ^3 - \dfrac{41}{6} \lambda ^2 + \dfrac{15}{2}\lambda -1 = 0
$$

となる。この方程式から固有値$\lambda _1$、$\lambda _2$、$\lambda _3$と、それぞれの固有ベクトル$x_1$、$x_2$、$x_3$が得られる。

$$
\begin{array}{}
\lambda_1 = 0.155、& \lambda_2 = 1.175、& \lambda_3 = 5.504\\
x_1 = 0.221、 & x_1 = 0.523、 & x_1 = 1.000\\
x_2 = 0.536、 & x_2 = 1.000、 & x_2 = -0.252\\
x_3 = 1.000、 & x_3 = -0.396、& x_3 = 0.016
\end{array}
$$

## 固有方程式

行列$\mathbb{B}$、$\mathbb{C}$は対称行列であっても、行列$\mathbb{A}$は一般的には対称行列にならない。

しかし、数学的扱いは$\mathbb{A}$が対称行列であることが望ましいので、式$(3)$の行列$\mathbb{A}$が対称行列になるような変換をしよう。

$\mathbb{C} = \mathbb{C}^{1/2} \mathbb{C}^{1/2}$ 、また、$\mathbb{C}^{1/2}\mathbb{C}^{-1/2} = \mathbb{I}$なので、式$(2)$は次のように変形できる。

$$\mathbb{B}(\mathbb{C}^{-1/2}\mathbb{C}^{1/2})\mathbb{x} = \lambda (\mathbb{C}^{1/2}\mathbb{C}^{1/2})\mathbb{x}$$

この両辺に左から逆行列$\mathbb{C}^{-1/2}$をかけると

$$\mathbb{C}^{-1/2}\mathbb{BC}^{-1/2}\mathbb{C}^{1/2}\mathbb{x} = \lambda \mathbb{C}^{1/2}\mathbb{x}\tag{4}$$

ここで、

$\mathbb{C}^{-1/2}\mathbb{BC}^{-1/2} = \mathbb{A}$、$\mathbb{C}^{1/2}\mathbb{x} = \mathbb{y}$

と置くと、

$$\mathbb{Ay} = \lambda \mathbb{y}$$

となる。ただし、$\mathbb{C}$が対称行列の場合に限られる。
